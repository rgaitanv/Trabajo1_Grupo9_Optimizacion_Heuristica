# 2D particulas Rosenbrock

```python
## Rastrigin 2D - optimización de partículas

import numpy as np
import matplotlib.pyplot as plt
from pyswarm import pso

# Definimos la función de Rastrigin
def rosenbrock_2d(x):
    return ((1-x[0])**2) + (100*(x[1]-(x[0]**2))**2)

# Definimos los límites de los parámetros
lb = np.array([-5.12, -5.12])
ub = np.array([5.12, 5.12])

# Optimizamos la función de Rastrigin utilizando el método de optimización de partículas
xopt, fopt = pso(rosenbrock_2d, lb, ub)

# Imprimimos los resultados
print("Punto óptimo: ", xopt)
print("Valor óptimo: ", fopt)

# Graficamos la función de Rastrigin y el punto óptimo
fig, ax = plt.subplots()
x = np.linspace(-5.12, 5.12, 100)
y = np.linspace(-5.12, 5.12, 100)
X, Y = np.meshgrid(x, y)
Z = rosenbrock_2d([X, Y])
ax.contourf(X, Y, Z, levels=100)
ax.set_title('Opt Particulas Rosenbrock 2D')
ax.plot(xopt[0], xopt[1], 'ro')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pyswarm import pso

def rosenbrock_3d(x):
    fx=((1-x[0])**2) + (100*(x[1]-(x[0]**2))**2) + ((1-x[1])**2) + (100*(x[2]-(x[1]**2))**2)
    return fx

# Definimos los límites de búsqueda
lb = [-2,-2,-2]
ub = [2,2,2]

# Ejecutamos la optimización de partículas usando PSO
xopt, fopt = pso(rosenbrock_3d, lb, ub, swarmsize=100, maxiter=100)

print("Punto óptimo: ", xopt)
print("Valor óptimo: ", fopt)
# Creamos una malla de puntos para la visualización de la función
x = np.arange(-2, 2, 0.05)
y = np.arange(-2, 2, 0.05)
X, Y = np.meshgrid(x, y)
Z = rosenbrock_3d([X, Y, np.arange(-2, 2, 0.05)])

# Creamos la figura y los ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficamos la función Rosenbrock 3D
ax.plot_surface(X, Y, Z, cmap='coolwarm')

# Graficamos la solución óptima encontrada por PSO
ax.scatter(xopt[0], xopt[1],xopt[2], color='red', marker='*', s=200)

# Configuramos los ejes y el título
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Opt Particulas Rosenbrock 3D')

# Mostramos la figura
plt.show()
```
