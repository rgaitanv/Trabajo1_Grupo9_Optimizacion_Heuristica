# Rosembrock 2D - Evolucion Diferencial

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import differential_evolution

# Definimos la función de Rastrigin
def rosenbrock_2d(x):
    fx=((1-x[0])**2) + (100*(x[1]-(x[0]**2))**2)
    return fx

# Definimos los límites de los parámetros
bounds = [(-5.12, 5.12), (-5.12, 5.12)]

# Optimizamos la función de Rastrigin utilizando el método de evolución diferencial
result = differential_evolution(rosenbrock_2d, bounds, strategy='best1bin', maxiter=1000, popsize=20, tol=0.01)

# Imprimimos los resultados
print("Punto óptimo: ", round(result.x[0],3),round(result.x[1],3))
print("Valor óptimo: ", result.fun)

# Graficamos la función de Rastrigin y el punto óptimo
fig, ax = plt.subplots()
x = np.linspace(-5.12, 5.12, 100)
y = np.linspace(-5.12, 5.12, 100)
X, Y = np.meshgrid(x, y)
Z = rosenbrock_2d([X, Y])
ax.contourf(X, Y, Z, levels=100)
ax.plot(result.x[0], result.x[1], 'ro')
plt.show()
```

# Rosembrock 3D - Evolucion Diferencial

```python
import numpy as np
from scipy.optimize import differential_evolution
import matplotlib.pyplot as plt

def rosenbrock_3d(x):
    fx=((1-x[0])**2) + (100*(x[1]-(x[0]**2))**2) + ((1-x[1])**2) + (100*(x[2]-(x[1]**2))**2)
    return fx

bounds = [(-5.12, 5.12) for _ in range(3)]
result = differential_evolution(rosenbrock_3d, bounds)

print("Best solution:",round(result.x[0],3),round(result.x[1],3),round(result.x[2],3))
print("Best fitness:", result.fun)


x = np.linspace(-5.12, 5.12, 100)
y = np.linspace(-5.12, 5.12, 100)
X, Y = np.meshgrid(x, y)
Z = rosenbrock_3d([X, Y, np.zeros_like(X)])

fig, ax = plt.subplots()
ax.contour(X, Y, Z, levels=np.logspace(0, 5, 35), cmap='jet')
ax.plot(result.x[0], result.x[1],result.x[2], 'r*', markersize=15)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Rosembrock 3D diferencial')
plt.show()
```
